```python
import numpy as np
import pandas as pd

from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import roc_auc_score
from sklearn.preprocessing import LabelEncoder

import lightgbm as lgb
import xgboost as xgb
from catboost import CatBoostClassifier

import warnings
warnings.filterwarnings("ignore")
```

```python
train = pd.read_csv("/content/train.csv")
test = pd.read_csv("/content/test.csv")
sample = pd.read_csv("/content/sample_submission.csv")

print(train.shape, test.shape)
train.head()
```

```python
TARGET = "y"
ID_COL = "id"
```

```python
X = train.drop([TARGET], axis=1)
y = train[TARGET]

X_test = test.copy()
```

```python
cat_cols = X.select_dtypes(include=["object"]).columns
num_cols = X.select_dtypes(exclude=["object"]).columns
```

```python
for col in cat_cols:
    le = LabelEncoder()
    all_vals = pd.concat([X[col], X_test[col]], axis=0).astype(str)
    le.fit(all_vals)

    X[col] = le.transform(X[col].astype(str))
    X_test[col] = le.transform(X_test[col].astype(str))
```

```python
N_FOLDS = 5
skf = StratifiedKFold(n_splits=N_FOLDS, shuffle=True, random_state=42)
```

```python
oof_lgb = np.zeros(len(X))
oof_xgb = np.zeros(len(X))
oof_cat = np.zeros(len(X))

test_lgb = np.zeros(len(X_test))
test_xgb = np.zeros(len(X_test))
test_cat = np.zeros(len(X_test))
```

```python
import lightgbm as lgb
from sklearn.metrics import roc_auc_score

oof_lgb = np.zeros(len(X))
test_lgb = np.zeros(len(X_test))

lgb_params = {
    "objective": "binary",
    "metric": "auc",
    "learning_rate": 0.03,
    "num_leaves": 64,
    "feature_fraction": 0.8,
    "bagging_fraction": 0.8,
    "bagging_freq": 5,
    "seed": 42
}

for fold, (train_idx, val_idx) in enumerate(skf.split(X, y)):
    print(f"LGB Fold {fold+1}")

    X_train, X_val = X.iloc[train_idx], X.iloc[val_idx]
    y_train, y_val = y.iloc[train_idx], y.iloc[val_idx]

    model = lgb.LGBMClassifier(
        **lgb_params,
        n_estimators=5000
    )

    model.fit(
        X_train,
        y_train,
        eval_set=[(X_val, y_val)],
        eval_metric="auc",
        callbacks=[
            lgb.early_stopping(stopping_rounds=200),
            lgb.log_evaluation(0)
        ]
    )

    oof_lgb[val_idx] = model.predict_proba(X_val)[:, 1]
    test_lgb += model.predict_proba(X_test)[:, 1] / N_FOLDS

print("‚úÖ LGB AUC:", roc_auc_score(y, oof_lgb))
```

```python
from catboost import CatBoostClassifier
from sklearn.metrics import roc_auc_score
import numpy as np

oof_cat = np.zeros(len(X))
test_cat = np.zeros(len(X_test))

for fold, (train_idx, val_idx) in enumerate(skf.split(X, y)):
    print(f"CAT Fold {fold+1}")

    X_train, X_val = X.iloc[train_idx], X.iloc[val_idx]
    y_train, y_val = y.iloc[train_idx], y.iloc[val_idx]

    model = CatBoostClassifier(
        iterations=800,           # ‚¨ÖÔ∏è FAST
        learning_rate=0.08,       # ‚¨ÖÔ∏è faster learning
        depth=6,                  # ‚¨ÖÔ∏è shallow = speed
        loss_function="Logloss",
        eval_metric="AUC",
        random_seed=42,
        verbose=100
    )

    model.fit(
        X_train,
        y_train,
        eval_set=(X_val, y_val),
        early_stopping_rounds=50
    )

    oof_cat[val_idx] = model.predict_proba(X_val)[:, 1]
    test_cat += model.predict_proba(X_test)[:, 1] / N_FOLDS

print("üî• CAT AUC:", roc_auc_score(y, oof_cat))
```

```python
oof_blend = 0.65 * oof_lgb + 0.35 * oof_cat
print("üöÄ BLEND AUC:", roc_auc_score(y, oof_blend))
```

```python
test_blend = 0.65 * test_lgb + 0.35 * test_cat
```

```python
submission = pd.DataFrame({
    "id": test["id"],
    "y": test_blend
})

submission.to_csv("submission.csv", index=False)
submission.head()
```
